```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import plot_tree
from sklearn.metrics import confusion_matrix, classification_report, mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split
```

## Task 1.1 - Trigonometry

```python
import math

class Trigonometry:

    def cos(self, x):
        return math.cos(x)

    def sin(self, x):
        return math.sin(x)

    def acos(self, x):
        return math.acos(x)

    def asin(self, x):
        return math.asin(x)

    def tan(self, x):
        return math.tan(x)

    def atan(self, x):
        return math.atan(x)

    def radians(self, degrees):
        return math.radians(degrees)

trig = Trigonometry()

print(trig.cos(math.pi/4))
```

## Task 1.3 - Decision tree

```python
X = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])
target = [0, 0, 0, 1, 1, 1]

X_test = np.array([[-2, -1], [-1, -1], [-2, -2], [1, 2], [2, 2], [3, 3]])
y_test = [0, 0, 0, 1, 1, 1]
```

```python
plt.scatter(X[:, 0], X[:, 1], c=target)
plt.show()
```

```python
classifier = DecisionTreeClassifier()
classifier.fit(X, target)
```

```python
plot_tree(classifier)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

## 1.4 - Decision Tree Regressor

```python
url = 'https://raw.githubusercontent.com/likarajo/petrol_consumption/master/data/petrol_consumption.csv'
dataset = pd.read_csv(url)
dataset.head()
```

```python
print(dataset.shape)
dataset.describe()
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
print(X)
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
regressor = DecisionTreeRegressor()
regressor.fit(X_train, y_train)
```

```python
plot_tree(regressor)
```

```python
y_pred = regressor.predict(X_test)
y_pred
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df
```

```python
print('MSE: ', mean_squared_error(y_test, y_pred))
print('MAN: ', mean_absolute_error(y_test, y_pred))
```
